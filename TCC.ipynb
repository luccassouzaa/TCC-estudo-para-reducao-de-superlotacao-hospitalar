```python
import pandas as pd
```

```python
base = pd.read_csv("dataset_hospitalar - dataset_fila_hospitalar.csv")
pd.set_option('display.max_columns', None)  # visualizar todo data set
```

```python
# quantas vezes um paciente visitou o hospital

base.groupby(['historico_visitas_ult_30d']).size()

# 15 pacientes "novos", onde nos ultimos 30 dias nao compareceram no hospital
```

```python
# desses pacientes, quais sao as principais prioridade_triagem e especialidade_necessaria

base.groupby(['prioridade_triagem','especialidade_necessaria']).size()
```

```python
# com isso também podemos calcular a media de atendimento em cada grupo
base["tempo_total"] = base["tempo_atendimento_medico_min"] + base["tempo_espera_atendimento_min"]

base.groupby(
    ["prioridade_triagem", "especialidade_necessaria"]
).agg(
    tempo_median=("tempo_total", "median"),
    quantidade=("tempo_total", "size")
).reset_index().sort_values(by="quantidade", ascending=False)
```

```python
# além disso, podemos ver também qual foi a media de cada atendimento em seu modo distinto (tempo_atendimento_medico_min e tempo_espera_atendimento_min)

base.groupby(
    ["prioridade_triagem", "especialidade_necessaria","tempo_atendimento_medico_min","tempo_espera_atendimento_min"]
).agg(
    # tempo_median=("tempo_total", "median"),
    quantidade=("tempo_total", "size")
).reset_index().sort_values(by="prioridade_triagem", ascending=False)
```

```python
# nesse momento, sabemos que ocupa mais tempo dentro de um hospital é quando um paciente passa por um clinico geral e é declarado como prioridade alta,
# vamos nos aprofundar nesses pacientes para ver se eles estão sendo diagnosticados corretamente

# base[(base['prioridade_triagem']=='Alta') & (base['especialidade_necessaria']=='Clínico Geral')]
```

```python
# desses 10 pacientes, podemos ver qual tipo de demanda foi feita para ele

base[(base['prioridade_triagem']=='Alta') & (base['especialidade_necessaria']=='Clínico Geral')].groupby('tipo_demanda').size()
```

```python
# 5 de acidente

base[
    (base['prioridade_triagem']=='Alta') & 
    (base['especialidade_necessaria']=='Clínico Geral') & 
    (base['tipo_demanda']=='Acidente')
].groupby('diagnostico_medico').size()
```

```python
# 2 de consulta (aqui podemos declarar como contusulta de rotina)

base[
    (base['prioridade_triagem']=='Alta') & 
    (base['especialidade_necessaria']=='Clínico Geral') & 
    (base['tipo_demanda']=='Consulta')
].groupby('diagnostico_medico').size()
```

```python
# 2 de retorno (provavelmente algum tipo de tratamento)

base[
    (base['prioridade_triagem']=='Alta') & 
    (base['especialidade_necessaria']=='Clínico Geral') & 
    (base['tipo_demanda']=='Retorno')
].groupby('diagnostico_medico').size()
```

```python
# 1 de transferencia

base[
    (base['prioridade_triagem']=='Alta') & 
    (base['especialidade_necessaria']=='Clínico Geral') & 
    (base['tipo_demanda']=='Transferência')
].groupby('diagnostico_medico').size()
```
